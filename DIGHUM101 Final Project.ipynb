### Removing Stopwords

```python
from nltk.corpus import stopwords
```

```python
def no_stops(song):
    # Removes stop words
    no_stops = [word for word in song if word not in stopwords.words('english')]
    return no_stops
```

```python
nipsey_1 = no_stops(nipsey_1)
nipsey_2 = no_stops(nipsey_2)
nipsey_3 = no_stops(nipsey_3)
schoolboy_1 = no_stops(schoolboy_1)
schoolboy_2 = no_stops(schoolboy_2)
schoolboy_3 = no_stops(schoolboy_3)
yg_1 = no_stops(yg_1)
yg_2 = no_stops(yg_2)
yg_3 = no_stops(yg_3)
jayrock_1 = no_stops(jayrock_1)
jayrock_2 = no_stops(jayrock_2)
jayrock_3 = no_stops(jayrock_3)
```

## Frequency Analysis

```python
from collections import Counter
```

```python
def frequency(song):
    # Determines the frequency of each non-stopword in the song
    freq = Counter(song)
    return freq.most_common()
```

```python
nipsey_1 = frequency(nipsey_1)
nipsey_2 = frequency(nipsey_2)
nipsey_3 = frequency(nipsey_3)
schoolboy_1 = frequency(schoolboy_1)
schoolboy_2 = frequency(schoolboy_2)
schoolboy_3 = frequency(schoolboy_3)
yg_1 = frequency(yg_1)
yg_2 = frequency(yg_2)
yg_3 = frequency(yg_3)
jayrock_1 = frequency(jayrock_1)
jayrock_2 = frequency(jayrock_2)
jayrock_3 = frequency(jayrock_3)
```

### Findings 

There were no clear distinctions between the two sides of the rag, with all of the artists emphasizing the pursuit of success, both financially and through public perception. The same indicators of status are used throughout the songs, with cars, jewelry, and other material goods used to reinforce their wealth. All of the artists similarly engage in acknowledgement of the trials and tribulations they have gone through in orfer to achieve what they have, noting the sacrifices made and the friends lost in that pursuit. However, I did notice a difference in the emphasis put on gang affiliation between the two sides, with the bloods generally being far more up front about their banging than the crips (pressing the message as opposed to simply recognizing it). The language used, too, was somewhat different, with Nipsey Hussle, ScHoolboy Q, and Jay Rock generally having longer, more complex words and lyricism when compared to YG. Words such as entrepeneurship, commercial, and enterprise appearing in Nipsey's music points to a particular economic focus when compared to the others, while ScHoolboy Q uses terms such as chromosome, fragrance, and kickstand, all indicate a higher mastery of rhyming schemes when compared to the blood/piru rappers. 

This analysis was somewhat biased by the inclusion of the songs' choruses in the dataset, as hip hop and rap choruses generally tend to be repetitive in terms of word usage. This naturally trended the word frequency values towards words in the choruses, as opposed to those within the actual verses. Similarly, more normalization of the data is needed (lemmitization perhaps) as many words were recounted due to inflection caused mispellings. For example, in Man of the Year, the word 'bounce' was counted differently from the word 'bbbbounce', due to the emphasis on the latter's first letter. 

My expectations were somewhat proven correct, though a sample of 3 songs is hardly enough by which to judge an artists thematic and lyrical spectrum. Crip rappers, Nipsey Hussle in particular, proved to engage in more diverse and complex topics when compared to their blood counterparts, who in turn followed the expecation of emphasizing their credentials more. As stated, however, 12 songs is not enough to make a definitve statement/conclusion.
